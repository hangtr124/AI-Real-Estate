```python
import sklearn
from sklearn import tree
from sklearn import preprocessing
from joblib import dump
import pandas as pd
import pathlib
```

```python
#Open the file and store the contents in the body variable
file = open('../data/synthetic_data_set.csv', 'r')
body = file.readlines()

#initialize the header and data_set list variables for storage
header = []
data_set = []

#split lines from the body of the data and store it in the data_set variable
for line in body:
    columns = line.split(',')
    if '\n' in columns:
        columns.remove('\n')
    data_set.append(columns)

#seperate the header from the data
header = data_set[0]
data_set.remove(data_set[0])

#print header first and then data set with space delimiters
#for x in header:
#    print(x.strip(), end=', ')
#print()
#for x in data_set:
#    for y in x:
#        print(y.strip(), end=', ')
#    print()
```

```python
print("Progress 0%")

#price, ppsf, total sq. ft., days to sell, zip code, bedrooms, bathrooms, half bathrooms, year built, garage,
#split the data_set list into features and labels lists
feature_columns = [[], [], [], [], [], [], [], [], []]#, []]
feature_set = []
label_columns = []#[], []]
label_set = []
#take each column and split it into the feature_set and label_set lists
for x in data_set:
    feature_columns[0].append(x[1])
    feature_columns[1].append(x[2])
    feature_columns[2].append(x[3])
    feature_columns[3].append(x[4])
    feature_columns[4].append(x[5])
    feature_columns[5].append(x[6])
    feature_columns[6].append(x[7])
    feature_columns[7].append(x[8])
    feature_columns[8].append(x[9])
    #feature_columns[9].append(x[11])
    label_columns.append(x[0].strip())
    #label_columns[1].append(x[3].strip())
    
print("Progress 10%")
      
#cast integer columns to int without encoding for features and labels
feature_columns[0] = [eval(x) for x in feature_columns[0]]
feature_columns[1] = [eval(x) for x in feature_columns[1]]
feature_columns[2] = [eval(x) for x in feature_columns[2]]
feature_columns[3] = [eval(x) for x in feature_columns[3]]
feature_columns[4] = [eval(x) for x in feature_columns[4]]
feature_columns[5] = [eval(x) for x in feature_columns[5]]
feature_columns[6] = [eval(x) for x in feature_columns[6]]
feature_columns[7] = [eval(x) for x in feature_columns[7]]
#feature_columns[8] = [eval(x) for x in feature_columns[8]]
label_columns = [eval(x) for x in label_columns]

print("Progress 20%")

#encoder declaration for preprocessing
encoder = preprocessing.LabelEncoder()
#encoded_listed = encoder.fit_transform(feature_columns[1])
encoded_garage = encoder.fit_transform(feature_columns[8])
#encoded_sold = encoder.fit_transform(label_columns[0])

print("Progress 30%")

#normalize the encoding into a list and reinsert it into their respective lists
temporary_list1 = []
temporary_list2 = []
temporary_list3 = []

#for x in encoded_listed: 
#    temporary_list1.append(x)
#feature_columns[1] = temporary_list1
for x in encoded_garage: 
    temporary_list2.append(x)
feature_columns[8] = temporary_list2
#for x in encoded_sold: 
#    temporary_list3.append(x)
#label_columns[0] = temporary_list3

print("Progress 40%")

for x in range(0, len(feature_columns[0])):
    feature_set.append([feature_columns[0][x], feature_columns[1][x], feature_columns[2][x], feature_columns[3][x], feature_columns[4][x], feature_columns[5][x],
                        feature_columns[6][x], feature_columns[7][x], feature_columns[8][x]])#, feature_columns[9][x]])
    
print("Progress 50%")
    
for x in range(0, len(label_columns)):
    label_set.append(label_columns[x])#, label_columns[1][x]])
    
print("Progress 80%")
    
#declare test and train sets for feature set and label set
test_feature_set = []
test_label_set = []
train_feature_set = []
train_label_set = []

#set aside 25% of the 20 data points in the data set
test_amount = int(len(feature_set) * .25)
for x in range(0, len(feature_set)):
    if x < test_amount:
        test_feature_set.append(feature_set[x])
        test_label_set.append(label_set[x])
    elif x >= test_amount:
        train_feature_set.append(feature_set[x])
        train_label_set.append(label_set[x])

print("Progress 90%")
        
#declare the learning model and fit to the model
decision_tree = tree.DecisionTreeClassifier()
decision_tree = decision_tree.fit(train_feature_set, train_label_set)

print("Progress 100%")

#print out the test label set to test the model
#print('Feature set from left to right is [Price, Listed, Zip, Bed, Bath, HalfBath, Sqrft, LotSqrft, Year, Garage]:')
#print(test_feature_set)
#print()
#print('Labels for the feature set from left to right is [Sold, Days]:')
#print(test_label_set)
```
